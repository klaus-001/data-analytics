<h2>(1) Define the Objective</h2>

The purpose of this notebook is to provide a visual and analytical representation of Netflixâ€™s content library in order to identify patterns, trends, and insights about content type, genre, geography, release patterns, and audience ratings.

<h2>(2) Collect & Understand the Data</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime as dt
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.info()
```

<h2>(3) Data Cleaning & Preparation</h2>

```python
df.isna().sum()
```

<h3>Dealing with the missing data</h3>

```python
# Replacements
df["country"] = df["country"].fillna(df["country"].mode()[0])

for column in ["director", "cast"]:
    df[column] = df[column].fillna("Not Available")

df["rating"] = df["rating"].fillna("PG")

# Drops
df.dropna(subset=["date_added"], inplace=True)

# Drop duplicates
df.drop_duplicates(inplace=True)
```

<h3>Missing values dealt with, but the date isn't quite right yet...</h3>

```python
df["date_added"] = df["date_added"].str.strip()
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")

df['month_added'] = df['date_added'].dt.month
df['month_name_added'] = df['date_added'].dt.month_name()
df['year_added'] = df['date_added'].dt.year
```

```python
df.info()
```

### Netflix through the years

Netflix started as DVD rentals, and now they have an audience of over 150m people - this is their story...

Timeline code from Subin An's awesome notebook   
https://www.kaggle.com/subinium/awesome-visualization-with-titanic-dataset

```python
# Main Events
main_events = [
    ("1997\nFounded", 1),
    ("1998\nMail Service", 2),
    ("2003\nGoes Public", 4),
    ("2007\nStreaming service", 5.3),
    ("2016\nGoes Global", 8),
    ("2021\nNetflix & Chill", 9),
]

# Sub Events
sub_events = [
    ("1998", "Netflix.com launched", 1.5),
    ("2000", "Starts\nPersonal\nRecommendations", 3),
    ("2006", "Billionth DVD Delivery", 5),
    ("2010", "Canadian\nLaunch", 6.5),
    ("2013", "Original\nContent", 7.2),
]

# Plot Setup
fig, ax = plt.subplots(figsize=(15, 4), constrained_layout=True)
ax.set_ylim(-2, 1.75)
ax.set_xlim(0, 10)

# Draw timeline base line
ax.axhline(0, xmin=0.1, xmax=0.9, color="#4a4a4a", zorder=1)

# Main Events (big circles + labels)
for label, x in main_events:
    ax.scatter(x, 0, s=120, c="#4a4a4a", zorder=2)
    ax.scatter(x, 0, s=30, c="#fafafa", zorder=3)
    ax.text(x, -0.55, label, ha="center", fontfamily="serif", fontweight="bold",
            color="#4a4a4a", fontsize=12)

# Sub Events (stems + labels)
levels = np.array([0.3 if i % 2 == 0 else -0.3 for i in range(len(sub_events))])
x_positions = [e[2] for e in sub_events]

# Draw stems
markerline, stemline, baseline = ax.stem(x_positions, levels)
plt.setp(baseline, zorder=0)
plt.setp(markerline, marker="o", color="#4a4a4a")
plt.setp(stemline, color="#4a4a4a")

# Add sub-event text
for idx, (year, desc, x) in enumerate(sub_events, start=1):
    color = "#b20710" if idx == len(sub_events) else "#4a4a4a"

    ax.text(x, 1.3*(idx % 2) - 0.5, year, ha="center", fontfamily="serif",
            fontweight="bold", color=color, fontsize=11)
    
    ax.text(x, 1.3*(idx % 2) - 0.6, desc, va="top", ha="center",
            fontfamily="serif", color=color)

# Remove spines and ticks
ax.set_xticks([])
ax.set_yticks([])
for spine in ["left", "top", "right", "bottom"]:
    ax.spines[spine].set_visible(False)

# Title & subtitle
ax.set_title("Netflix through the years",
             fontweight="bold", fontfamily="serif", fontsize=16, color="#4a4a4a")
ax.text(2.4, 1.57,
        "From DVD rentals to a global audience of over 150m people - is it time for Netflix to Chill?",
        fontfamily="serif", fontsize=12, color="#4a4a4a")

plt.show()
```

### By Country

So we now know there are much more movies than TV shows on Netflix (which surprises me!).

What about if we look at content by country?

I would imagine that the USA will have the most content. I wonder how other countries will compare?

```python
# Make a copy & explode countries
df_country = df.copy()
df_country['country'] = df_country['country'].str.split(', ')
df_country = df_country.explode('country')

# Helper column (1 per row, for counting)
df_country['count'] = 1

# Shorten long country names
df_country['country'] = df_country['country'].replace({
    'South Korea': 'S. Korea',
    'United States': 'USA',
    'United Kingdom': 'UK'
})

# Aggregate top 10 countries
data = (
    df_country.groupby('country')['count']
    .sum()
    .sort_values(ascending=False)[:10]
)
```

```python
# Create color mapping
color_map = ['#f5f5f1'] * 10   # light gray for all
color_map[0:3] = ['#b20710'] * 3  # highlight top 3 in Netflix red

# Plot
fig, ax = plt.subplots(figsize=(12, 6))

bars = ax.bar(
    data.index, data,
    width=0.5,
    edgecolor='darkgray',
    linewidth=0.6,
    color=color_map
)

# Annotations (values above bars)
offset = data.max() * 0.05  # 5% above bar height
for idx, val in enumerate(data):
    ax.annotate(
        f"{val}",
        xy=(idx, val + offset),
        ha='center', va='center',
        fontweight='light', fontfamily='serif'
    )

# Style the plot
for s in ['top', 'left', 'right']: # Remove borders except bottom
    ax.spines[s].set_visible(False)

# Keep labels but remove tick marks
ax.set_xticks(range(len(data)))
ax.set_xticklabels(data.index, fontfamily='serif', rotation=0)
ax.tick_params(axis='both', which='both', length=0, labelsize=12)

# Add horizontal gridlines
ax.grid(axis='y', linestyle='-', alpha=0.4)
ax.set_axisbelow(True)

# Custom y ticks
ax.set_yticks(np.arange(0, 4000, 500))

# Thicken bottom line
plt.axhline(y=0, color='black', linewidth=1.3, alpha=0.7)

# Titles and Insights
fig.text(0.09, 1, 'Top 10 Countries on Netflix',
         fontsize=15, fontweight='bold', fontfamily='serif')
fig.text(0.09, 0.95, 'The three most frequent countries have been highlighted.',
         fontsize=12, fontweight='light', fontfamily='serif')

fig.text(1.05, 0.9, 'Insight',
         fontsize=15, fontweight='bold', fontfamily='serif')

fig.text(1.05, 0.5,
         '''The most prolific producers of
content for Netflix are, primarily,
the USA, with India and the UK
a significant distance behind.

It makes sense that the USA produces 
the most content as, afterall, 
Netflix is a US company.''',
         fontsize=12, fontweight='light', fontfamily='serif')

# Add a vertical divider line (optional design element)
import matplotlib.lines as lines
l1 = lines.Line2D([1, 1], [0, 1],
                  transform=fig.transFigure, figure=fig,
                  color='black', lw=0.2)
fig.lines.extend([l1])

plt.show()
```

As predicted, the USA dominates.

The UK is a top contender too, but still some way behind India.

How does content by country vary?

```python
# Prepare the data
country_order = df_country['country'].value_counts().head(10).index

# Count how many Movies vs TV Shows per country
country_type_counts = (
    df_country.groupby(['country', 'type']).size()
    .unstack(fill_value=0)
    .loc[country_order]
)

# Add total for each country
country_type_counts['total'] = country_type_counts.sum(axis=1)

# Convert counts to ratios (percentages)
country_ratios = (
    (country_type_counts.T / country_type_counts['total']).T[['Movie', 'TV Show']]
)
country_ratios = country_ratios.sort_values("Movie", ascending=False)[::-1]
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

# Plot stacked horizontal bars
ax.barh(country_ratios.index, country_ratios['Movie'],
        color='#b20710', alpha=0.8, label='Movie')
ax.barh(country_ratios.index, country_ratios['TV Show'],
        left=country_ratios['Movie'],
        color='#221f1f', alpha=0.8, label='TV Show')

# Clean axis
ax.set_xlim(0, 1)
ax.set_xticks([])
ax.set_yticks(range(len(country_ratios.index)))
ax.set_yticklabels(country_ratios.index, fontfamily='serif', fontsize=11)

# Annotate percentages inside the bars
for country in country_ratios.index:
    movie_pct = country_ratios['Movie'][country] * 100
    tv_pct = country_ratios['TV Show'][country] * 100
    
    ax.annotate(f"{movie_pct:.1f}%", 
                xy=(country_ratios['Movie'][country] / 2, country),
                va='center', ha='center', color='white',
                fontsize=11, fontfamily='serif')
    
    ax.annotate(f"{tv_pct:.1f}%", 
                xy=(country_ratios['Movie'][country] + country_ratios['TV Show'][country] / 2, country),
                va='center', ha='center', color='white',
                fontsize=11, fontfamily='serif')

# Titles and labels
fig.text(0.12, 0.94, 'Top 10 Countries: Movie vs TV Show Split',
         fontsize=15, fontweight='bold', fontfamily='serif')
fig.text(0.12, 0.91, 'Percent stacked bar chart',
         fontsize=12, fontfamily='serif')

# Legend styled as text
fig.text(0.75, 0.93, 'Movie', fontweight="bold", fontfamily='serif', fontsize=13, color='#b20710')
fig.text(0.82, 0.93, '|', fontweight="bold", fontfamily='serif', fontsize=13, color='black')
fig.text(0.84, 0.93, 'TV Show', fontweight="bold", fontfamily='serif', fontsize=13, color='#221f1f')

# Insights box
fig.text(1.05, 0.94, 'Insight', fontsize=15, fontweight='bold', fontfamily='serif')
fig.text(1.05, 0.45,
         '''India is dominated by Movies, reflecting
Bollywood's global influence.

South Korea, on the other hand, leans
heavily towards TV Shows, driven by
the global success of K-dramas.

These differences highlight how Netflix
adapts its content strategy by region.''',
         fontsize=12, fontfamily='serif', fontweight='light')

import matplotlib.lines as lines

# Create a vertical line (x = 1 across the figure, from y=0 to y=1)
l1 = lines.Line2D(
    [1, 1],        # x-coordinates in figure space
    [0, 1],        # y-coordinates in figure space
    transform=fig.transFigure,  # use figure coordinates, not data coords
    figure=fig,
    color='black',
    lw=0.2         # line width
)

# Add it to the figure
fig.lines.extend([l1])

# Remove plot borders
for s in ['top', 'left', 'right', 'bottom']:
    ax.spines[s].set_visible(False)

# Remove tick marks but keep labels
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```

As I've noted in the insights on the plot, it is really interesting to see how the split of TV Shows and Movies varies by country.

South Korea is dominated by TV Shows - why is this? I am a huge fan of South Korean cinema so I know they have a great movie selection.

Equally, India is dominated by Movies. I think this might be due to Bollywood - comment below if you have any other ideas!

### Ratings

Let's briefly check out how ratings are distributed

```python
df["count"] = 1 # helper column for plotting

# Order ratings by total frequency
rating_order_df = (
    df.groupby("rating")["count"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

rating_order = list(rating_order_df["rating"])

# Create pivot table (Movie vs TV Show counts by rating)
rating_split = (
    df.groupby("type")["rating"]
    .value_counts()
    .unstack()
    .sort_index()
    .fillna(0)
    .astype(int)[rating_order] # reindex by ordered ratings
)

# Separate Movie and TV counts
movie_counts = rating_split.loc["Movie"]
tv_counts = -rating_split.loc["TV Show"] # negative for downward bars
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Bars
ax.bar(movie_counts.index, movie_counts, width=0.5,
       color="#b20710", alpha=0.8, label="Movie")
ax.bar(tv_counts.index, tv_counts, width=0.5,
       color="#221f1f", alpha=0.8, label="TV Show")

# Annotations - TV Show (below baseline)
for rating in tv_counts.index:
    ax.annotate(f"{-tv_counts[rating]}",
                xy=(rating, tv_counts[rating] - 60),
                va="center", ha="center",
                fontweight="light", fontfamily="serif",
                color="#4a4a4a")

# Annotations - Movie (above baseline)
for rating in tv_counts.index:
    ax.annotate(f"{movie_counts[rating]}",
                xy=(rating, movie_counts[rating] + 60),
                va="center", ha="center",
                fontweight="light", fontfamily="serif",
                color="#4a4a4a")

# Style plot
for spine in ["top", "left", "right", "bottom"]:
    ax.spines[spine].set_visible(False)

ax.set_xticks(range(len(rating_split.columns)))
ax.set_xticklabels(rating_split.columns, fontfamily="serif")
ax.set_yticks([])

ax.legend().set_visible(False)

# Add title and description
fig.text(0.16, 1, 'Rating distribution by Film & TV Show',
         fontsize=15, fontweight='bold', fontfamily='serif')
fig.text(0.16, 0.89,
'''We observe that some ratings are only applicable to Movies. 
The most common for both Movies & TV Shows are TV-MA and TV-14.
''',
         fontsize=12, fontweight='light', fontfamily='serif')

# Custom legend with text
fig.text(0.755, 0.924, "Movie", fontweight="bold",
         fontfamily='serif', fontsize=15, color='#b20710')
fig.text(0.815, 0.924, "|", fontweight="bold",
         fontfamily='serif', fontsize=15, color='black')
fig.text(0.825, 0.924, "TV Show", fontweight="bold",
         fontfamily='serif', fontsize=15, color='#221f1f')

plt.show()
```

### How has content been added over the years?

As we saw in the timeline at the start of this analysis, Netflix went global in 2016 - and it is extremely noticeable in this plot.

The increase in Movie content is remarkable.

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Define colors for each type
colors = {
    "Movie": "#b20710",   # Netflix red
    "TV Show": "#221f1f"  # Netflix black
}

# Loop through each type (Movie / TV Show)
for show_type in df["type"].unique():
    # Count titles added per year for that type
    yearly_counts = (
        df.loc[df["type"] == show_type, "year_added"]
          .value_counts()
          .sort_index()
    )

    # Plot line and shaded area
    ax.plot(yearly_counts.index, yearly_counts, color=colors[show_type])
    ax.fill_between(
        yearly_counts.index, 0, yearly_counts, 
        color=colors[show_type], alpha=0.9
    )

# Add baseline line
ax.axhline(y=0, color="black", linewidth=1.3, alpha=0.7)

# Remove spines
for spine in ["top", "right", "bottom", "left"]:
    ax.spines[spine].set_visible(False)

# Format axes
ax.set_xlim(2008, 2020)
ax.set_xticks(np.arange(2008, 2021, 1))
ax.tick_params(axis="both", length=0)   # Remove tick marks

# Show y-axis ticks on the right
ax.yaxis.tick_right()

# Titles & Explanations
fig.text(
    0.13, 0.85, "Movies & TV Shows added over time", 
    fontsize=15, fontweight="bold", fontfamily="serif"
)

fig.text(
    0.13, 0.59,
    """We see a slow start for Netflix over several years. 
Things begin to pick up in 2015 and then there is a 
rapid increase from 2016.

It looks like content additions have slowed down in 2020, 
likely due to the COVID-19 pandemic.""",
    fontsize=12, fontweight="light", fontfamily="serif"
)

# Custom Legend
fig.text(
    0.13, 0.2, "Movie", 
    fontweight="bold", fontfamily="serif", fontsize=15, color=colors["Movie"]
)
fig.text(
    0.19, 0.2, "|", 
    fontweight="bold", fontfamily="serif", fontsize=15, color="black"
)
fig.text(
    0.2, 0.2, "TV Show", 
    fontweight="bold", fontfamily="serif", fontsize=15, color=colors["TV Show"]
)

plt.show()
```

### We can view the same plot, but as a cumulative total...

```python
# Prepare cumulative data by year and type
data_sub = (
    df.groupby("type")["year_added"]
      .value_counts()
      .unstack()
      .fillna(0)
      .loc[["TV Show", "Movie"]]
      .cumsum(axis=0)   # cumulative total of movies and tv shows per year
      .T
)

# Create figure
fig, ax = plt.subplots(figsize=(12, 6))
colors = ["#b20710", "#221f1f"]  # Movie = red, TV Show = black

# Plot cumulative fill for Movies and TV Shows
for i, content_type in enumerate(df["type"].value_counts().index):
    ax.fill_between(
        data_sub.index,
        0,
        data_sub[content_type],
        color=colors[i],
        alpha=0.9
    )

# Axis styling
ax.yaxis.tick_right()
ax.axhline(y=0, color="black", linewidth=1.3, alpha=0.7)
ax.set_xlim(2008, 2020)
ax.set_xticks(np.arange(2008, 2021, 1))
ax.tick_params(axis="both", which="both", length=0)

# Remove borders
for spine in ["top", "right", "bottom", "left"]:
    ax.spines[spine].set_visible(False)

# Titles & descriptive text
fig.text(0.13, 0.85,
         "Movies & TV Shows added over time [Cumulative Total]",
         fontsize=15, fontweight="bold", fontfamily="serif")

fig.text(0.13, 0.58,
         "Netflix peak global content amount was in 2019.\n\n"
         "It appears that Netflix has focused more attention\n"
         "on increasing Movie content than TV Shows.\n"
         "Movies have increased much more dramatically\n"
         "than TV shows.",
         fontsize=12, fontweight="light", fontfamily="serif")

# Custom legend (inline text instead of plt.legend)
fig.text(0.13, 0.2, "Movie", fontweight="bold",
         fontfamily="serif", fontsize=15, color="#b20710")
fig.text(0.19, 0.2, "|", fontweight="bold",
         fontfamily="serif", fontsize=15, color="black")
fig.text(0.2, 0.2, "TV Show", fontweight="bold",
         fontfamily="serif", fontsize=15, color="#221f1f")

plt.show()
```

### Month-by-Month

We've seen how content has increased over the years, but are there certain months that, on average, tend to enjoy more content being added?

I'll show this in a couple of ways - a cumulative year view, and also as a radial plot...

```python
month_order = [
    'January',
    'February',
    'March',
    'April',
    'May',
    'June',
    'July',
    'August',
    'September',
    'October',
    'November',
    'December'
]

df["month_name_added"] = pd.Categorical(df["month_name_added"],
                                       categories=month_order,
                                       ordered=True)
```

```python
# Prepare cumulative monthly content by type
data_sub = (
    df.groupby("type")["month_name_added"]
      .value_counts()
      .unstack(fill_value=0)
      .loc[["TV Show", "Movie"]]  # ensure ordering
      .cumsum(axis=0)             # cumulative sum over months
      .T                          # transpose â†’ months as index
)

# Setup figure and axes
fig, ax = plt.subplots(figsize=(12, 6))
colors = ["#b20710", "#221f1f"]

# --- Plot filled areas for Movies and TV Shows
for i, content_type in enumerate(df["type"].value_counts().index):
    ax.fill_between(
        data_sub.index,
        0,
        data_sub[content_type],
        color=colors[i],
        label=content_type,
        alpha=0.9
    )

# Style and formatting
for side in ["top", "right", "bottom", "left"]:
    ax.spines[side].set_visible(False)
    
ax.grid(False)
ax.yaxis.tick_right()
ax.axhline(y=0, color="black", linewidth=1.3, alpha=0.4)

# X-axis formatting
ax.set_xticks(range(len(month_order)))
ax.set_xticklabels(data_sub.index, fontfamily="serif", rotation=0)
ax.margins(x=0)
ax.tick_params(axis="both", which="both", length=0)

# Hide legend (we annotate manually instead)
ax.legend().set_visible(False)

# Title and text annotations
fig.text(0.13, 0.95,
         "Content added by month [Cumulative Total]",
         fontsize=15, fontweight="bold", fontfamily="serif")

fig.text(0.13, 0.905,
         "The end & beginnings of each year seem to be Netflix's preference for adding content.",
         fontsize=12, fontweight="light", fontfamily="serif")

# Custom manual legend
fig.text(0.13, 0.855, "Movie", fontsize=15, fontweight="bold",
         fontfamily="serif", color="#b20710")
fig.text(0.19, 0.855, "|", fontsize=15, fontweight="bold",
         fontfamily="serif", color="black")
fig.text(0.2, 0.855, "TV Show", fontsize=15, fontweight="bold",
         fontfamily="serif", color="#221f1f")

plt.show()
```

### What about a more interesting way to view how content is added across the year?

Sometimes visualizations should be eye-catching & attention grabbing - I think this visual acheives that, even if it isn't the most precise.

By highlighting certain months, the reader's eye is drawn exactly where we want it.

```python
def plot_polar_chart(data_sub, save_as=None):
    """
    Creates a polar bar chart of Netflix content added by month.

    Parameters
    ----------
    data_sub : pd.DataFrame
        DataFrame containing at least ['month_name_added', 'Movie', 'TV Show'].
    save_as : str or None
        File path (e.g., 'chart.png') to save the figure. If None, only displays the chart.
    """

    # --- Prepare data ---
    data_sub2 = data_sub.copy()
    data_sub2["Value"] = data_sub2["Movie"] + data_sub2["TV Show"]
    data_sub2 = data_sub2.reset_index()

    # Sort by month order (descending for polar layout)
    df_polar = data_sub2.sort_values(by="month_name_added", ascending=False)

    # Color map: all black, Jan + Dec highlighted red
    color_map = ["#221f1f" for _ in range(12)]
    color_map[0] = color_map[11] = "#b20710"

    # --- Create polar chart ---
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)
    plt.axis("off")

    # Layout constants
    upperLimit = 30
    lowerLimit = 1
    labelPadding = 30

    # Normalize values to fit within limits
    max_val = df_polar["Value"].max()
    slope = (max_val - lowerLimit) / max_val
    heights = slope * df_polar["Value"] + lowerLimit

    # Bar width and angle positions
    width = 2 * np.pi / len(df_polar.index)
    indexes = range(1, len(df_polar.index) + 1)
    angles = [i * width for i in indexes]

    # Draw bars
    bars = ax.bar(
        x=angles,
        height=heights,
        width=width,
        bottom=lowerLimit,
        linewidth=2,
        edgecolor="white",
        color=color_map,
        alpha=0.8,
    )

    # Add labels
    for bar, angle, height, label in zip(bars, angles, heights, df_polar["month_name_added"]):
        rotation = np.rad2deg(angle)

        # Flip labels for readability
        if np.pi / 2 <= angle < 3 * np.pi / 2:
            alignment = "right"
            rotation += 180
        else:
            alignment = "left"

        ax.text(
            x=angle,
            y=lowerLimit + bar.get_height() + labelPadding,
            s=label,
            ha=alignment,
            va="center",
            fontsize=10,
            fontfamily="serif",
            rotation=rotation,
            rotation_mode="anchor",
        )

    # --- Save or show ---
    if save_as:
        plt.savefig(save_as, dpi=300, bbox_inches="tight")
    plt.show()
```

```python
plot_polar_chart(data_sub)
```

Yes, December & January are definitely the best months for new content. Maybe Netflix knows that people have a lot of time off from work over this period and that it is a good time to reel people in?

February is the worst - why might this be? Ideas welcomed!

### Movie Genres

Let's now explore movie genres a little...

```python
from sklearn.preprocessing import MultiLabelBinarizer 
import matplotlib.colors
import seaborn as sns

# Custom colour map based on Netflix palette
cmap = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ['#221f1f', '#b20710', '#f5f5f1']
)

def genre_heatmap(df, title):
    # âœ… Always work on a safe copy
    df = df.copy()

    # Clean and split genres into lists
    df['genre'] = (
        df['listed_in']
        .str.replace(' ,', ',', regex=False)
        .str.replace(', ', ',', regex=False)
        .str.split(',')
    ) 

    # Collect all unique genres
    all_genres = set(g for sublist in df['genre'] for g in sublist)

    print(f"There are {len(all_genres)} types in the Netflix {title} Dataset")    

    # Convert multi-label lists into a binary matrix
    mlb = MultiLabelBinarizer()
    res = pd.DataFrame(mlb.fit_transform(df['genre']), 
                       columns=mlb.classes_, 
                       index=df.index)

    # Compute correlation
    corr = res.corr()

    # Build upper triangle mask
    mask = np.zeros_like(corr, dtype=bool)
    mask[np.triu_indices_from(mask)] = True

    # --- Plot
    fig, ax = plt.subplots(figsize=(10, 7))
    fig.text(.54, .88, 'Genre correlation', fontfamily='serif',
             fontweight='bold', fontsize=15)
    fig.text(.75, .665,
             '''
             It is interesting that Independent Movies
             tend to be Dramas. 
             
             Another observation is that 
             International Movies are rarely
             in the Children's genre.
             ''',
             fontfamily='serif', fontsize=12, ha='right')

    sns.heatmap(
        corr, mask=mask, cmap=cmap,
        vmax=.3, vmin=-.3, center=0,
        square=True, linewidths=2.5
    )

    plt.show()
```

```python
df_tv = df[df["type"] == "TV Show"]
df_movies = df[df["type"] == "Movie"]

genre_heatmap(df_movies, 'Movie')
```

### Let's have a quick look at the lag between when content is released and when it is added on Netflix

Spain looks to have a lot of new content. Great for them!

```python
# We would be using this dataframes from earlier to create the next visuals
# df_movies
# df_tv
```

```python
def plot_avg_age_by_country(df, country_col, title, subtitle, color_release="#221f1f", color_added="#b20710"):
    """
    Plot average release vs added year for top 10 countries in dataset.
    
    Parameters
    ----------
    df : pd.DataFrame
        Input dataframe (e.g. df_movies or df_tv).
    country_col : str
        Column name for country grouping (e.g. 'country' or 'first_country').
    title : str
        Title text for the figure.
    subtitle : str
        Commentary/annotation text for the figure.
    color_release : str
        Color for release year scatter.
    color_added : str
        Color for added year scatter.
    """

    # Top 10 countries
    top_countries = (
        df.groupby(country_col)["count"]
        .sum()
        .reset_index()
        .nlargest(10, "count")
    )

    # Filter data
    df_top = df[df[country_col].isin(top_countries[country_col])].copy()

    # Compute averages
    avg_years = (
        df_top.groupby(country_col)[["release_year", "year_added"]]
        .mean()
        .round()
    )

    # Sort by release year
    avg_years_sorted = avg_years.sort_values(by="release_year")
    my_range = range(1, len(avg_years_sorted.index) + 1)

    # Plot
    fig, ax = plt.subplots(figsize=(7, 5))

    # Title
    fig.text(0.13, 0.9, title, fontsize=15, fontweight="bold", fontfamily="serif")

    # Horizontal lines
    plt.hlines(
        y=my_range,
        xmin=avg_years_sorted["release_year"],
        xmax=avg_years_sorted["year_added"],
        color="grey", alpha=0.4
    )

    # Scatter points
    plt.scatter(
        avg_years_sorted["release_year"], my_range,
        color=color_release, s=100, alpha=0.9, label="Average release date"
    )
    plt.scatter(
        avg_years_sorted["year_added"], my_range,
        color=color_added, s=100, alpha=0.9, label="Average added date"
    )

    # Style cleanup
    for spine in ["top", "left", "right", "bottom"]:
        ax.spines[spine].set_visible(False)

    ax.yaxis.tick_right()
    ax.tick_params(axis="both", which="both", length=0)
    plt.yticks(my_range, avg_years_sorted.index, fontname="serif", fontsize=12)

    # Custom legend
    fig.text(0.19, 0.175, "Released", fontweight="bold", fontfamily="serif", fontsize=12, color=color_release)
    fig.text(0.47, 0.175, "Added", fontweight="bold", fontfamily="serif", fontsize=12, color=color_added)

    # Commentary
    fig.text(0.13, 0.42, subtitle, fontsize=12, fontweight="light", fontfamily="serif")

    plt.show()
```

```python
plot_avg_age_by_country(
    df_movies,
    country_col="country",
    title="How old are the movies? [Average]",
    subtitle="""The average gap between when 
content is released, and when it
is then added on Netflix varies
by country. 

In Spain, Netflix appears to be 
dominated by newer movies 
whereas Egypt & India have
an older average movie."""
)
```

What about TV shows...

```python
plot_avg_age_by_country(
    df_tv,
    country_col="country",
    title="How old are the TV shows? [Average]",
    subtitle="""The gap for TV shows seems
more regular than for movies.

This is likely due to subsequent
series being released
year-on-year.

Spain seems to have
the newest content overall."""
)
```

### USA & India

As the two largest content countries, it might be fun to compare the two

```python
# Filter USA & India data
us_ind = df_country[df_country["country"].isin(["USA", "India"])]

# Build cumulative dataset for plotting
data_sub = (
    df_country.groupby("country")["year_added"]
    .value_counts()
    .unstack(fill_value=0)
    .loc[["USA", "India"]]
    .cumsum(axis=1)  # cumulative over years
    .T
)
```

```python
# Plot
fig, ax = plt.subplots(figsize=(12, 6))
colors = {"USA": "#221f1f", "India": "#b20710"}

for country in ["USA", "India"]:
    yearly_counts = (
        us_ind[us_ind["country"] == country]["year_added"]
        .value_counts()
        .sort_index()
    )
    ax.fill_between(
        yearly_counts.index,
        0,
        yearly_counts,
        color=colors[country],
        alpha=0.9,
        label=country
    )
    ax.plot(yearly_counts.index, yearly_counts, color=colors[country])

# Styling
ax.set_ylim(0, 1000)
ax.yaxis.tick_right()
ax.axhline(y=0, color="black", linewidth=1.3, alpha=0.4)

# Hide spines
for spine in ["top", "right", "bottom", "left"]:
    ax.spines[spine].set_visible(False)

# Ticks and labels
ax.grid(False)
ax.set_xlim(2008, 2020)
ax.set_xticks(np.arange(2008, 2022, 1))
ax.set_xticklabels(data_sub.index, fontfamily="serif", rotation=0)
ax.margins(x=0)
ax.tick_params(axis="both", which="both", length=0)

# Text annotations
fig.text(0.13, 0.85, "USA vs. India: When was content added?",
         fontsize=15, fontweight="bold", fontfamily="serif")

fig.text(0.13, 0.58,
"""We know from our work above that Netflix is dominated by the USA & India.
It would also be reasonable to assume that, since Netflix is an American
company, Netflix increased content first in the USA, before 
other nations. 

That is exactly what we see here; a slow and then rapid
increase in content for the USA, followed by Netflix 
being launched to the Indian market in 2016.""",
fontsize=12, fontweight="light", fontfamily="serif")

# Custom legend-like labels
fig.text(0.13, 0.15, "India", fontweight="bold", fontfamily="serif",
         fontsize=15, color=colors["India"])
fig.text(0.19, 0.15, "|", fontweight="bold", fontfamily="serif",
         fontsize=15, color="black")
fig.text(0.20, 0.15, "USA", fontweight="bold", fontfamily="serif",
         fontsize=15, color=colors["USA"])

plt.show()
```

So the USA dominates. But is there a plot that can convey this in another way?

```python
us_ind = df_country[(df_country['country'] == 'USA') | (df_country['country'] == 'India' )]

data_sub = df_country.groupby('country')['year_added'].value_counts().unstack().fillna(0).loc[['USA','India']].cumsum(axis=0).T
data_sub.insert(0, "base", np.zeros(len(data_sub)))

data_sub = data_sub.add(-us_ind['year_added'].value_counts()/2, axis=0)

# --- Prepare Stream Graph for USA vs India ---
fig, ax = plt.subplots(figsize=(14, 6))

# Colors: India (red), USA (black) -> reversed order for layering
colors = ['#221f1f', '#b20710']
countries = data_sub.columns  # ['USA', 'India']

# Build the stacked areas
for i in range(1, len(countries)):
    ax.fill_between(
        data_sub.index,
        data_sub.iloc[:, i-1],
        data_sub.iloc[:, i],
        color=colors[i-1]
    )

# --- Style and Formatting ---
for side in ['top', 'right', 'bottom', 'left']:
    ax.spines[side].set_visible(False)

ax.set_axisbelow(True)
ax.set_yticks([])
ax.grid(False)
ax.tick_params(axis='y', which='both', length=0)

# --- Titles & Annotations ---
fig.text(0.16, 0.76, 'USA vs. India: Stream graph of new content added',
         fontsize=15, fontweight='bold', fontfamily='serif')

fig.text(0.16, 0.575, 
'''
Seeing the data displayed like this helps 
us to realise just how much content is added in the USA.
Remember, India has the second largest amount of
content yet is dwarfed by the USA.'''
, fontsize=12, fontweight='light', fontfamily='serif')

# Custom legend-like labels
fig.text(0.16, 0.41, "India", fontweight="bold", fontfamily='serif',
         fontsize=15, color='#b20710')
fig.text(0.208, 0.41, "|", fontweight="bold", fontfamily='serif',
         fontsize=15, color='black')
fig.text(0.218, 0.41, "USA", fontweight="bold", fontfamily='serif',
         fontsize=15, color='#221f1f')

plt.show()
```

### Lastly, we can view a wordcloud to get an overview of Netflix titles

It is interesting to note that many films share the same key words in their titles.

Credit to Dmitry Uarov for figuring this visual out. His notebook is here:

```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.colors

# Custom colour map
cmap = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ['#221f1f', '#b20710']
)

# Prepare text
text = " ".join(df['title'].astype(str))

# Load mask and invert it
mask = np.array(Image.open("netflix_logo.png"))
mask_inverted = np.invert(mask)  # flip black/white

# Create WordCloud with inverted mask
wordcloud = WordCloud(
    background_color="white",
    mask=mask_inverted,
    colormap=cmap,
    max_words=150,
    contour_color="white",
    contour_width=1
).generate(text)

# Display
plt.figure(figsize=(6,6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
